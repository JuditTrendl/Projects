```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import numpy as np # linear algebra
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\Users\Csucsu\Desktop\data.csv")
```

```python
data.head()
```

```python
def roletype(type):
    if type[1]=="N" and type[2]=="T":
        return "Analyst"
    elif type[1]=="N" and type[2]=="F":
        return "Diplomat"
    elif type[1]=="S" and type[3]=="J":
        return "Sentinel"
    elif type[1]=="S" and type[3]=="P":
        return "Explorer"
    else:
        return "Error"
data['role'] = data['type'].apply(lambda x: roletype(x))
```

```python
df1 = data[data['role'] == 'Diplomat']
df2 = data[data['role'] == 'Analyst']
df3 = data[data['role'] == 'Sentinel']
df4 = data[data['role'] == 'Explorer']
```

```python
stopwords = set(STOPWORDS)
stopwords.add('youtube')
stopwords.add('https')
stopwords.add('ESTP')
stopwords.add('ISTP')
stopwords.add('INFJ')
stopwords.add('ESFJ')
stopwords.add('ISFP')
stopwords.add('ENTP')
stopwords.add('INTP')
stopwords.add('INFP')
stopwords.add('really')
stopwords.add('watch')
stopwords.add("don'")
```

```python
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(df1['posts']))
```

```python
fig = plt.figure(1,figsize=(10,10));
plt.imshow(wordcloud);
plt.axis('off');
plt.title('Diplomat',size=20);
plt.show();
plt.savefig('Diplomat');
```
